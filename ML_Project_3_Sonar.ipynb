# Model Selection

```python
# Classification with 2 classes
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
import warnings
warnings.filterwarnings("ignore")
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=50)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
model=[LogisticRegression(), DecisionTreeClassifier(criterion='gini'), DecisionTreeClassifier(criterion='entropy'),
      SVC(kernel='poly'), SVC(kernel='rbf'), SVC(kernel='linear'), KNeighborsClassifier(), GaussianNB()]
```

```python
for m in model:
    m.fit(x_train, y_train)
    y=m.predict(x_test)
    print(accuracy_score(y,y_test))
```

```python
# max score is corresponding to SVC(rbf)
```

```python
for i in range(5,20):
    knn=KNeighborsClassifier(n_neighbors=i)
    knn.fit(x_train,y_train)
    y=knn.predict(x_test)
    print(i, " ", accuracy_score(y,y_test))
```

```python
y.shape
```

```python
max_score=0
r_state=0
for i in range(45,150):
    x1_train, x1_test, y1_train, y1_test=train_test_split(x,y,test_size=0.2, random_state=i)
    svc=SVC(kernel='rbf')
    svc.fit(x1_train, y1_train)
    y_pred=svc.predict(x1_test)
    acc=accuracy_score(y_pred,y1_test)
    print(i, "  ", acc)
    if acc > max_score:
        max_score=acc
        r_state=i
print("acc ", max_score, " ", "r_state ", r_state)
```

```python
# thus, the right r_state is 63
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=63)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
for m in model:
    m.fit(x_train, y_train)
    y_pred=m.predict(x_test)
    print(accuracy_score(y_pred,y_test))
```

```python
# still, best is svc(rbf)
```

# Cross Validation

```python
from sklearn.model_selection import cross_val_score, cross_val_predict
```

```python
max_sc=0
cfold=0
for i in range(3,111):
    score=cross_val_score(SVC(kernel='rbf'), x,y,cv=i)
    print(i, " ",score.mean(), " ", score.std())
    if score.mean() > max_sc:
        max_sc=score.mean()
        cfold=i
print(cfold, " ", max_sc)
```

```python
# the cfold should be 110
```

```python
score=cross_val_score(SVC(kernel='rbf'), x,y,cv=110)
print(score.mean(), " ", score.std())
pred=cross_val_predict(SVC(kernel='rbf'), x,y,cv=110)
print(pred)
print(accuracy_score(y,pred))
```

```python
# there is not much problem of over fitting or under fitting
```

```python
for m in model:
    score=cross_val_score(m,x,y,cv=110)
    print(score.mean(), " ", score.std())
    pred=cross_val_predict(m,x,y,cv=110)
    print(accuracy_score(y,pred))
```

```python
svc=SVC()
g=GridSearchCV(svc, param_grid,cv=110)
```

# Ensembling

```python
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
```

```python
rf=RandomForestClassifier(n_estimators=50)
```

```python
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
accuracy_score(pred,y_test)
```

```python
rf.score(x_train, y_train)
```

```python
for i in range(1,200,10):
    rf=RandomForestClassifier(n_estimators=i)
    rf.fit(x_train,y_train)
    pred=rf.predict(x_test)
    print(i, " ", accuracy_score(pred,y_test))
    print(rf.score(x_train, y_train))
```

```python
abc=AdaBoostClassifier(n_estimators=70, base_estimator=SVC(kernel='rbf'), algorithm='SAMME')
```

```python
abc.fit(x_train, y_train)
print(abc.score(x_train, y_train))
pred=abc.predict(x_test)
print(accuracy_score(pred,y_test))
```

# AUC_ROC Curve

```python
from sklearn.metrics import roc_curve, roc_auc_score
```

```python
svc=SVC(kernel='rbf', probability=True)
svc.fit(x_train,y_train)
```

```python
y_pred_pro=svc.predict_proba(x_test)[:,1]
y_pred_pro
```

```python
fpr,tpr,threshold=roc_curve(y_test,y_pred_pro)
print(fpr)
print(tpr)
print(threshold)
```

```python
plt.plot([0,1], [0,1], 'k--')
plt.plot(fpr,tpr,label='SVC')
plt.show()
```

```python
auc_score=roc_auc_score(y_test,svc.predict(x_test))
print(auc_score)
```
